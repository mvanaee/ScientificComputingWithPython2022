```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])
print('1. using outer:\n' ,np.outer(u,v),'\n')
a = np.zeros((len(u),len(v)), dtype = int)
for i, u_i in enumerate(u):
    for j, v_j in enumerate(v):
        a[i][j] = u_i * v_j
print('2. using for loop:\n',a,'\n')
print('3. using broadcast:\n', u[:,np.newaxis] * v,'\n')
```

```python
import math

def prime_numbers(N):
    mask = np.ones(N+1, dtype=bool)
    mask[0] = False
    mask[1] = False
    for i in range(2,int(math.sqrt(N))+1):
        if mask[i]:
            for j in range(i**2, N+1, i):
                mask[j] = False
    
    numbers = np.arange(N+1)
    primes = numbers[mask]
    return primes
            
print(prime_numbers(1000), '\n')

%timeit prime_numbers(100)
%timeit prime_numbers(1000)
%timeit prime_numbers(10000)

# As it is shown, as N grows, time it takes to calculate prime numbers grows hugely.
#And it was quite expectable, becuase of iteration.

print('\nAs it is shown, as N grows, time it takes to calculate prime numbers grows hugely.')
print('And it was quite expectable, becuase of iteration.')
```
